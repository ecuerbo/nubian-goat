```python
import pandas as pd
import seaborn as sns
import numpy as np

df = pd.read_csv('/Users/ecuerbo/python/pyprac/data/ph-poverty.csv')
```

```python
df[['REG_NUM','NAME']] =df['Region'].str.split('(', expand=True)
df['NAME'] = df['NAME'].str.strip(')')
df.head()
```

```python
df['REG_NUM'].unique()
```

```python
df['NAME'].unique()
```

```python
cols=['Region', 'Region code', 'Province', 'Province code',
            'annual_poverty_thresh_2006',
            'annual_poverty_thresh_2009',
            'annual_poverty_thresh_2012',
            'annual_poverty_thresh_2015',
            'poverty_incidence_est_2006',
            'poverty_incidence_est_2009',
            'poverty_incidence_est_2012',
            'poverty_incidence_est_2015',
            'poverty_incidence_cv_2006',
            'poverty_incidence_cv_2009',
            'poverty_incidence_cv_2012',
            'poverty_incidence_cv_2015',
            'poor_fam_2006','poor_fam_2009','poor_fam_2012','poor_fam_2015',
             'REG_NUM','NAME']
#Annual Per Capita Poverty Threshold = annual_poverty_thresh_
#'Poverty Incidence among Families (%)_Estimated (%)= poverty_incidence_est_
#'Poverty Incidence among Families (%)_CV (%)= poverty_incidence_cv_
# Magnitude of Poor Families_Est_ = poor_fam_
df.columns = cols
df
```

```python
df1= pd.wide_to_long(df,stubnames=['annual_poverty_thresh','poverty_incidence_est','poverty_incidence_cv','poor_fam'],
                         i=['Region', 'Province'], j='year', sep='_').reset_index()
df1.drop(['Region','Province code','Region code'],axis=1,inplace=True)
```

```python
df1['NAME'] = df1['NAME'].replace({'NCR':'NATIONAL CAPITAL REGION',"CAR":'CORDILLERA ADMINISTRATIVE REGION','ARMM':'AUTONOMOUS REGION IN MUSLIM MINDANAO'}).str.upper()
df1['REG_NUM'] = df1['REG_NUM'].replace({'NATIONAL CAPITAL REGION ':'NCR','CORDILLERA ADMINISTRATIVE REGION ':'CAR','AUTONOMOUS REGION IN MUSLIM MINDANAO':'ARMM'})
df1['NAME'] = df1['NAME'].fillna('MIMAROPA')
df1['annual_poverty_thresh'].replace(',',"",regex=True,inplace=True)
df1['poor_fam'].replace(',',"",regex=True,inplace=True)
df1['annual_poverty_thresh'] = df1['annual_poverty_thresh'].astype('int')
#df1['poor_fam'] = df1['poor_fam'].astype('int32')
df1.head()
```
